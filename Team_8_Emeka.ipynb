```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Extracting relevant columns for linear regression
X = merged_data[['AcresBurned']]  # Independent variable
y = merged_data['TotalGeneration']  # Dependent variable

# Initialize the Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Predict the total generation based on wildfire acres burned
y_pred = model.predict(X)

# Display the regression coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficient: {model.coef_}")

# Create the plot with enhanced aesthetics
plt.figure(figsize=(12, 8))

# Scatter plot of actual data with transparency and black edges
plt.scatter(X, y, color='orange', label='Actual Data', alpha=0.7, edgecolors='black', s=100)

# Plot the regression line with a thicker, more noticeable line
plt.plot(X, y_pred, color='dodgerblue', linewidth=3, label='Regression Line')

# Title and labels with improved fonts
plt.title('Linear Regression: Acres Burned vs Energy Generation (California)', fontsize=18, weight='bold')
plt.xlabel('Acres Burned', fontsize=14)
plt.ylabel('Total Generation (GWh)', fontsize=14)

# Customize the gridlines
plt.grid(True, which='both', linestyle='--', linewidth=0.6)

# Set background color and style for the plot
plt.gca().set_facecolor('whitesmoke')

# Customize the legend
plt.legend(fontsize=12, loc='upper left')

# Tight layout to ensure everything fits nicely
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Assuming 'merged_data' is already available and cleaned
# Step 1: Calculate the "Severity" of the wildfire (just based on Acres Burned in this case)
merged_data['Severity'] = merged_data['AcresBurned']  # Using Acres Burned as severity

# Step 2: Prepare the data for linear regression
# The independent variable (X) is "Severity" (Acres Burned) and the dependent variable (y) is "TotalGeneration" (energy loss).
X_severity = merged_data[['Severity']]  # Independent variable
y_loss = merged_data['TotalGeneration']  # Dependent variable

# Step 3: Initialize and fit the linear regression model
model_severity = LinearRegression()
model_severity.fit(X_severity, y_loss)

# Step 4: Plot the regression line
plt.figure(figsize=(10, 6))
plt.scatter(X_severity, y_loss, color='orange', label='Energy loss data')
plt.plot(X_severity, model_severity.predict(X_severity), color='blue', label='Regression Line')
plt.xlabel('Wildfire Severity (Acres Burned)')
plt.ylabel('Energy Loss (GWh)')
plt.title('Impact of Wildfire Severity on Energy Loss')
plt.legend()
plt.grid(True)
plt.show()

# Print model coefficients (optional)
print(f"Intercept: {model_severity.intercept_}")
print(f"Coefficient: {model_severity.coef_}")
```
